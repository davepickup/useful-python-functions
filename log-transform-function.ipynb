```python
#Log transform skewed features

import numpy as np
import pandas as pd
import scipy.stats as stats
from scipy.stats import skew

def log_transform(dataframe, num_data, threshold):
    skewed_feats = dataframe[list(num_data)].apply(lambda x: skew(x.dropna()))
    skewed_feats = skewed_feats[skewed_feats > threshold]
    skewed_feats = skewed_feats.index
    dataframe[skewed_feats] = np.log1p(dataframe[skewed_feats])
    return dataframe
```
